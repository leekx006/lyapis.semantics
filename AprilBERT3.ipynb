```python
import os
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from transformers import AutoModel, AutoTokenizer
import numpy as np
from tqdm import tqdm
import pandas as pd
import matplotlib.pyplot as plt
import random
import pickle
from google.colab import drive

# First, mount Google Drive
drive.mount('/content/drive')

# Set up paths
BASE_PATH = '/content/drive/MyDrive/latin_model'
MODEL_DIR = f'{BASE_PATH}/contrastive_model'
DATA_DIR = '/content/drive/MyDrive/OHG-texts'  # Updated path with lowercase 't'

# Create model directory if it doesn't exist
os.makedirs(MODEL_DIR, exist_ok=True)

print(f"Model will be saved to: {MODEL_DIR}")

# Device configuration
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")

# PART 1: MODEL SAVING AND LOADING UTILITIES
def save_model_robust(model, path, save_full_model=False):
    """
    Save model with multiple fallbacks and verification

    Args:
        model: The PyTorch model to save
        path: Path where to save the model
        save_full_model: If True, save the entire model, not just state_dict

    Returns:
        bool: True if saving was successful, False otherwise
    """
    print(f"Attempting to save model to {path}...")

    # Ensure directory exists
    os.makedirs(os.path.dirname(path), exist_ok=True)

    # Try multiple saving methods
    methods_tried = 0
    max_attempts = 3

    # Method 1: Standard PyTorch save
    try:
        methods_tried += 1
        print(f"Method {methods_tried}/{max_attempts}: Standard PyTorch save")
        if save_full_model:
            torch.save(model, path)
        else:
            torch.save(model.state_dict(), path)

        # Verify file exists and has content
        if os.path.exists(path) and os.path.getsize(path) > 0:
            size_mb = os.path.getsize(path) / (1024 * 1024)
            print(f"✓ Success! Saved to {path} ({size_mb:.2f} MB)")

            # Force sync with Google Drive if in Colab
            try:
                print("Syncing with Google Drive...")
                drive.flush_and_unmount()
                import time
                time.sleep(2)  # Wait for flush to complete
                drive.mount('/content/drive')

                # Verify file still exists after remounting
                if os.path.exists(path) and os.path.getsize(path) > 0:
                    size_mb = os.path.getsize(path) / (1024 * 1024)
                    print(f"✓ Verified after Drive sync: {path} ({size_mb:.2f} MB)")
                    return True
                else:
                    print(f"✗ File missing after Drive sync: {path}")
            except Exception as e:
                print(f"Drive sync warning (not critical): {e}")
                return True  # Still return True as the save worked
        else:
            print(f"✗ File verification failed for {path}")
    except Exception as e:
        print(f"✗ Method {methods_tried} failed: {e}")

    # Method 2: Save with pickle
    try:
        methods_tried += 1
        print(f"Method {methods_tried}/{max_attempts}: Pickle")
        pickle_path = f"{path}.pickle"
        with open(pickle_path, 'wb') as f:
            if save_full_model:
                pickle.dump(model, f)
            else:
                pickle.dump(model.state_dict(), f)

        if os.path.exists(pickle_path) and os.path.getsize(pickle_path) > 0:
            size_mb = os.path.getsize(pickle_path) / (1024 * 1024)
            print(f"✓ Success! Saved to {pickle_path} ({size_mb:.2f} MB)")
            return True
        else:
            print(f"✗ File verification failed for {pickle_path}")
    except Exception as e:
        print(f"✗ Method {methods_tried} failed: {e}")

    # Method 3: Save to temp location and copy
    try:
        methods_tried += 1
        print(f"Method {methods_tried}/{max_attempts}: Temp file + system copy")
        import random
        temp_path = f"/tmp/model_temp_{random.randint(1000, 9999)}.pt"

        if save_full_model:
            torch.save(model, temp_path)
        else:
            torch.save(model.state_dict(), temp_path)

        # Use system commands to copy file
        os.system(f"cp {temp_path} {path}")

        if os.path.exists(path) and os.path.getsize(path) > 0:
            size_mb = os.path.getsize(path) / (1024 * 1024)
            print(f"✓ Success! Saved to {path} ({size_mb:.2f} MB)")
            return True
        else:
            print(f"✗ File verification failed for {path}")
    except Exception as e:
        print(f"✗ Method {methods_tried} failed: {e}")

    print(f"All {max_attempts} saving methods failed. Could not save model to {path}")
    return False

def load_model_robust(model_class, path, device=None):
    """
    Load a model with multiple fallbacks and helpful error messages

    Args:
        model_class: The model class to instantiate
        path: Path to the saved model
        device: Device to load the model to (None for auto-detection)

    Returns:
        The loaded model, or None if loading failed
    """
    if device is None:
        device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

    print(f"Attempting to load model from {path} to {device}...")

    # Method 1: Try direct loading of state_dict
    if os.path.exists(path):
        try:
            model = model_class().to(device)
            model.load_state_dict(torch.load(path, map_location=device))
            print(f"✓ Successfully loaded model state_dict from {path}")
            return model
        except Exception as e:
            print(f"✗ Error loading state_dict: {e}")
    else:
        print(f"✗ Path does not exist: {path}")

    # Method 2: Try full model loading
    if os.path.exists(path):
        try:
            model = torch.load(path, map_location=device)
            print(f"✓ Successfully loaded full model from {path}")
            return model
        except Exception as e:
            print(f"✗ Error loading full model: {e}")

    # Method 3: Try pickle format
    pickle_path = f"{path}.pickle"
    if os.path.exists(pickle_path):
        try:
            with open(pickle_path, 'rb') as f:
                state_dict = pickle.load(f)

            model = model_class().to(device)
            model.load_state_dict(state_dict)
            print(f"✓ Successfully loaded model from pickle: {pickle_path}")
            return model
        except Exception as e:
            print(f"✗ Error loading from pickle: {e}")
    else:
        print(f"✗ Pickle path does not exist: {pickle_path}")

    # Method 4: Try to find any model file in the directory
    dir_path = os.path.dirname(path)
    if os.path.exists(dir_path):
        try:
            model_files = [f for f in os.listdir(dir_path) if f.endswith('.pt') or f.endswith('.pickle')]
            if model_files:
                print(f"Found alternative model files in the directory: {model_files}")
                print("You might try loading one of these instead.")
        except Exception:
            pass

    print("All loading methods failed.")
    return None

# PART 2: MODEL DEFINITION

# Define the contrastive model
class ContrastiveModel(nn.Module):
    def __init__(self, base_model_name='bert-base-multilingual-cased'):
        super(ContrastiveModel, self).__init__()
        self.bert = AutoModel.from_pretrained(base_model_name)

        # Projection layers - match the architecture we determined earlier
        self.projection = nn.Sequential(
            nn.Linear(768, 768),
            nn.ReLU(),
            nn.Linear(768, 128)
        )

    def forward(self, input_ids, attention_mask):
        outputs = self.bert(input_ids=input_ids, attention_mask=attention_mask)
        cls_output = outputs.last_hidden_state[:, 0, :]
        projection = self.projection(cls_output)
        projection = nn.functional.normalize(projection, p=2, dim=1)
        return projection

# PART 3: DATASET DEFINITION

# Dataset class for parallel texts
class ParallelTextDataset(Dataset):
    def __init__(self, latin_texts, ohg_texts, tokenizer, max_length=128):
        self.latin_texts = latin_texts
        self.ohg_texts = ohg_texts
        self.tokenizer = tokenizer
        self.max_length = max_length

    def __len__(self):
        return len(self.latin_texts)

    def __getitem__(self, idx):
        latin = self.latin_texts[idx]
        ohg = self.ohg_texts[idx]

        latin_encoding = self.tokenizer(
            latin,
            return_tensors='pt',
            max_length=self.max_length,
            padding='max_length',
            truncation=True
        )

        ohg_encoding = self.tokenizer(
            ohg,
            return_tensors='pt',
            max_length=self.max_length,
            padding='max_length',
            truncation=True
        )

        return {
            'latin_input_ids': latin_encoding['input_ids'].squeeze(),
            'latin_attention_mask': latin_encoding['attention_mask'].squeeze(),
            'ohg_input_ids': ohg_encoding['input_ids'].squeeze(),
            'ohg_attention_mask': ohg_encoding['attention_mask'].squeeze()
        }

# PART 4: LOSS FUNCTION

# Contrastive loss function
class ContrastiveLoss(nn.Module):
    def __init__(self, temperature=0.5):
        super(ContrastiveLoss, self).__init__()
        self.temperature = temperature

    def forward(self, z_i, z_j):
        batch_size = z_i.size(0)

        # Compute similarity matrix
        z = torch.cat([z_i, z_j], dim=0)
        sim = torch.mm(z, z.t()) / self.temperature

        # Mask for positive pairs
        sim_i_j = torch.diag(sim, batch_size)
        sim_j_i = torch.diag(sim, -batch_size)

        positive_samples = torch.cat([sim_i_j, sim_j_i], dim=0).reshape(2 * batch_size, 1)

        # Create mask for self-similarity (we don't want to include these in loss)
        mask = torch.ones_like(sim)
        mask = mask.fill_diagonal_(0)

        # Create numerator and denominator
        neg_samples = sim[mask.bool()].reshape(2 * batch_size, -1)

        # Compute loss
        negatives = torch.cat([neg_samples], dim=1)

        loss = -torch.log(
            torch.exp(positive_samples) /
            (torch.exp(positive_samples) + torch.sum(torch.exp(negatives), dim=1, keepdim=True))
        )

        return loss.mean()

# PART 5: HELPER FUNCTIONS

# Function to create negative pairs for training
def create_negative_pairs(texts, n_pairs=None):
    """Create negative pairs by pairing texts randomly"""
    n = len(texts)
    if n_pairs is None:
        n_pairs = n

    negative_pairs = []
    for _ in range(n_pairs):
        idx1 = random.randint(0, n-1)
        idx2 = random.randint(0, n-1)

        # Ensure the indices are different
        while idx1 == idx2:
            idx2 = random.randint(0, n-1)

        negative_pairs.append((texts[idx1], texts[idx2]))

    return negative_pairs

# PART 6: IMPROVED TRAINING FUNCTION

def train_model(model, train_loader, val_loader, epochs=5, lr=1e-5, checkpoint_every=100):
    """Train the contrastive model with frequent checkpointing and reliable saving"""
    optimizer = torch.optim.Adam(model.parameters(), lr=lr)
    criterion = ContrastiveLoss()

    best_val_loss = float('inf')
    train_losses = []
    val_losses = []

    # Track global batch count
    global_batch = 0

    # Safety check - ensure model directory exists
    os.makedirs(MODEL_DIR, exist_ok=True)

    # Function to verify saved file exists and has size > 0
    def verify_saved_file(path):
        if os.path.exists(path):
            size_mb = os.path.getsize(path) / (1024 * 1024)
            if size_mb > 0:
                print(f"✓ Verified: {path} ({size_mb:.2f} MB)")
                return True
            else:
                print(f"✗ File exists but is empty: {path}")
                return False
        else:
            print(f"✗ File does not exist: {path}")
            return False

    # Check if we can resume from a partial epoch
    resume_path = os.path.join(MODEL_DIR, 'partial_checkpoint.pt')
    start_epoch = 0

    if os.path.exists(resume_path):
        print(f"Resuming from checkpoint: {resume_path}")
        try:
            checkpoint = torch.load(resume_path)
            model.load_state_dict(checkpoint['model_state_dict'])
            optimizer.load_state_dict(checkpoint['optimizer_state_dict'])
            start_epoch = checkpoint['epoch']
            global_batch = checkpoint['global_batch']
            best_val_loss = checkpoint['best_val_loss']

            if 'train_losses' in checkpoint:
                train_losses = checkpoint['train_losses']
                val_losses = checkpoint['val_losses']

            print(f"Resuming from epoch {start_epoch}, batch {global_batch}")
        except Exception as e:
            print(f"Error loading checkpoint: {e}")
            print("Starting from scratch")

    for epoch in range(start_epoch, epochs):
        # Training
        model.train()
        epoch_loss = 0
        batch_count = 0

        progress_bar = tqdm(train_loader, desc=f"Epoch {epoch+1}/{epochs}")
        for batch_idx, batch in enumerate(progress_bar):
            latin_input_ids = batch['latin_input_ids'].to(device)
            latin_attention_mask = batch['latin_attention_mask'].to(device)
            ohg_input_ids = batch['ohg_input_ids'].to(device)
            ohg_attention_mask = batch['ohg_attention_mask'].to(device)

            # Forward pass
            latin_embeddings = model(latin_input_ids, latin_attention_mask)
            ohg_embeddings = model(ohg_input_ids, ohg_attention_mask)

            # Compute loss
            loss = criterion(latin_embeddings, ohg_embeddings)

            # Backward and optimize
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()

            epoch_loss += loss.item()
            batch_count += 1
            global_batch += 1

            progress_bar.set_postfix({'loss': loss.item(), 'batch': global_batch})

            # Save checkpoint every N batches
            if global_batch % checkpoint_every == 0:
                checkpoint = {
                    'epoch': epoch,
                    'global_batch': global_batch,
                    'model_state_dict': model.state_dict(),
                    'optimizer_state_dict': optimizer.state_dict(),
                    'loss': loss.item(),
                    'best_val_loss': best_val_loss,
                    'train_losses': train_losses,
                    'val_losses': val_losses
                }

                # Save with verification
                checkpoint_path = os.path.join(MODEL_DIR, f'checkpoint_batch_{global_batch}.pt')
                save_model_robust(checkpoint, checkpoint_path, save_full_model=True)

                # Also update the partial checkpoint
                save_model_robust(checkpoint, resume_path, save_full_model=True)
                print(f"\nSaved checkpoint at batch {global_batch}")

        # Compute average training loss for this epoch
        avg_train_loss = epoch_loss / batch_count
        train_losses.append(avg_train_loss)

        # Validation
        model.eval()
        val_loss = 0
        val_batch_count = 0

        with torch.no_grad():
            for batch in val_loader:
                latin_input_ids = batch['latin_input_ids'].to(device)
                latin_attention_mask = batch['latin_attention_mask'].to(device)
                ohg_input_ids = batch['ohg_input_ids'].to(device)
                ohg_attention_mask = batch['ohg_attention_mask'].to(device)

                # Forward pass
                latin_embeddings = model(latin_input_ids, latin_attention_mask)
                ohg_embeddings = model(ohg_input_ids, ohg_attention_mask)

                # Compute loss
                loss = criterion(latin_embeddings, ohg_embeddings)
                val_loss += loss.item()
                val_batch_count += 1

        # Compute average validation loss
        avg_val_loss = val_loss / val_batch_count
        val_losses.append(avg_val_loss)

        print(f"Epoch {epoch+1}/{epochs}, Train loss: {avg_train_loss:.4f}, Val loss: {avg_val_loss:.4f}")

        # Save best model
        if avg_val_loss < best_val_loss:
            best_val_loss = avg_val_loss

            # Save in multiple formats to ensure it works
            best_model_path = os.path.join(MODEL_DIR, 'best_contrastive_model.pt')
            save_model_robust(model, best_model_path)

            # Also save a backup for this epoch
            backup_path = os.path.join(MODEL_DIR, f'best_model_epoch_{epoch+1}.pt')
            save_model_robust(model, backup_path)

        # Save epoch checkpoint
        checkpoint = {
            'epoch': epoch + 1,  # Save the next epoch to resume from
            'global_batch': global_batch,
            'model_state_dict': model.state_dict(),
            'optimizer_state_dict': optimizer.state_dict(),
            'best_val_loss': best_val_loss,
            'train_losses': train_losses,
            'val_losses': val_losses
        }
        epoch_checkpoint_path = os.path.join(MODEL_DIR, f'checkpoint_epoch_{epoch+1}.pt')
        save_model_robust(checkpoint, epoch_checkpoint_path, save_full_model=True)

    # Save final model
    final_model_path = os.path.join(MODEL_DIR, 'final_contrastive_model.pt')
    save_model_robust(model, final_model_path)

    # Plot training curve
    plt.figure(figsize=(10, 5))
    plt.plot(train_losses, label='Training Loss')
    plt.plot(val_losses, label='Validation Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.title('Training and Validation Loss')
    plt.legend()
    plt.savefig(os.path.join(MODEL_DIR, 'training_curve.png'))
    plt.savefig(os.path.join(BASE_PATH, 'training_curve.png'))  # Save to parent dir as well

    return train_losses, val_losses

# PART 7: MAIN TRAINING FUNCTION

def train_contrastive_model():
    # Load tokenizer
    tokenizer = AutoTokenizer.from_pretrained('bert-base-multilingual-cased')

    # Load data from Excel files
    def load_excel_aligned_texts(excel_path, source_name):
        """Extract parallel texts from Excel files with aligned Latin and OHG"""
        try:
            print(f"Loading {source_name} from {excel_path}...")

            # Read the Excel file
            df = pd.read_excel(excel_path)

            # Use specific column names
            latin_col = 'Source clause'
            ohg_col = 'Target clause'

            print(f"  Using columns: '{latin_col}' (Latin) and '{ohg_col}' (OHG)")

            # Extract text pairs, skipping empty or null rows
            latin_texts = []
            ohg_texts = []

            for _, row in df.iterrows():
                latin = str(row[latin_col]).strip()
                ohg = str(row[ohg_col]).strip()

                # Skip rows where either text is empty, NaN, or \N
                if (latin and ohg and
                    latin != 'nan' and ohg != 'nan' and
                    latin != '\\N' and ohg != '\\N'):
                    latin_texts.append(latin)
                    ohg_texts.append(ohg)

            print(f"  Extracted {len(latin_texts)} valid pairs from {source_name}")
            return latin_texts, ohg_texts

        except Exception as e:
            print(f"Error processing {source_name}: {e}")
            return [], []

    # Load data from Excel files
    tatian_latin, tatian_ohg = load_excel_aligned_texts(
        os.path.join(DATA_DIR, 'Tatian_aligned.xlsx'), 'Tatian')

    isidor_latin, isidor_ohg = load_excel_aligned_texts(
        os.path.join(DATA_DIR, 'Isidor_aligned.xlsx'), 'Isidor')

    physio_latin, physio_ohg = load_excel_aligned_texts(
        os.path.join(DATA_DIR, 'Physiologus_aligned.xlsx'), 'Physiologus')

    # Combine all texts
    all_latin = tatian_latin + isidor_latin + physio_latin
    all_ohg = tatian_ohg + isidor_ohg + physio_ohg

    print(f"Total loaded: {len(all_latin)} parallel texts for training")

    # Create negative pairs
    negative_pairs = create_negative_pairs(all_latin, len(all_latin))
    print(f"Created {len(negative_pairs)} negative pairs")

    # Combine positive and negative pairs for training
    # Positive pairs
    train_data = []
    for latin, ohg in zip(all_latin, all_ohg):
        train_data.append((latin, ohg, 1))  # 1 indicates positive pair

    # Add negative pairs
    for latin1, latin2 in negative_pairs:
        train_data.append((latin1, latin2, 0))  # 0 indicates negative pair

    # Shuffle the data
    random.shuffle(train_data)

    # Split into training and validation sets (90/10 split)
    split_idx = int(0.9 * len(train_data))
    train_data, val_data = train_data[:split_idx], train_data[split_idx:]

    print(f"Training set: {len(train_data)} examples")
    print(f"Validation set: {len(val_data)} examples")

    # Function to convert data to dataset format
    def prepare_dataset(data_list, tokenizer):
        latin_texts = []
        ohg_texts = []

        for latin, ohg, _ in data_list:
            latin_texts.append(latin)
            ohg_texts.append(ohg)

        return ParallelTextDataset(latin_texts, ohg_texts, tokenizer)

    # Create datasets and dataloaders
    train_dataset = prepare_dataset(train_data, tokenizer)
    val_dataset = prepare_dataset(val_data, tokenizer)

    # Use a smaller batch size for GPU memory efficiency
    batch_size = 16
    train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
    val_loader = DataLoader(val_dataset, batch_size=batch_size)

    # Initialize model
    model = ContrastiveModel().to(device)

    # Train the model
    train_losses, val_losses = train_model(
        model,
        train_loader,
        val_loader,
        epochs=5,
        checkpoint_every=100  # Save checkpoint every 100 batches
    )

    # Verify saved model files
    print("\nChecking saved model files:")
    best_model_path = os.path.join(MODEL_DIR, 'best_contrastive_model.pt')
    final_model_path = os.path.join(MODEL_DIR, 'final_contrastive_model.pt')

    if os.path.exists(best_model_path):
        size_mb = os.path.getsize(best_model_path) / (1024 * 1024)
        print(f"✓ {best_model_path} exists ({size_mb:.2f} MB)")
    else:
        print(f"✗ {best_model_path} does not exist")

    if os.path.exists(final_model_path):
        size_mb = os.path.getsize(final_model_path) / (1024 * 1024)
        print(f"✓ {final_model_path} exists ({size_mb:.2f} MB)")
    else:
        print(f"✗ {final_model_path} does not exist")

    # Test the trained model
    print("\nTesting trained model:")
    test_latin = "Quoniam quidem multi conati sunt"
    test_ohg = "Bithiu uuanta manage zilotun"

    # Load the best model for testing
    test_model = ContrastiveModel().to(device)
    loaded_model = load_model_robust(ContrastiveModel, best_model_path, device)

    if loaded_model is not None:
        test_model = loaded_model
    else:
        print("Could not load best model, using final model instead")
        loaded_model = load_model_robust(ContrastiveModel, final_model_path, device)
        if loaded_model is not None:
            test_model = loaded_model
        else:
            print("Warning: Using untrained model for testing!")

    test_model.eval()

    # Tokenize
    latin_encoding = tokenizer(
        test_latin,
        return_tensors='pt',
        max_length=128,
        padding='max_length',
        truncation=True
    ).to(device)

    ohg_encoding = tokenizer(
        test_ohg,
        return_tensors='pt',
        max_length=128,
        padding='max_length',
        truncation=True
    ).to(device)

    # Compute embeddings
    with torch.no_grad():
        latin_embedding = test_model(latin_encoding['input_ids'], latin_encoding['attention_mask'])
        ohg_embedding = test_model(ohg_encoding['input_ids'], ohg_encoding['attention_mask'])

    # Compute similarity
    similarity = torch.nn.functional.cosine_similarity(latin_embedding, ohg_embedding).item()
    print(f"\nLatin: {test_latin}")
    print(f"OHG: {test_ohg}")
    print(f"Similarity: {similarity:.4f}")

    # Test with another example
    test_latin = "in principio erat verbum"
    test_ohg = "thaz uuir thar uuizzin thuruh mina minna iouuiht suachen"

    # Tokenize
    latin_encoding = tokenizer(
        test_latin,
        return_tensors='pt',
        max_length=128,
        padding='max_length',
        truncation=True
    ).to(device)

    ohg_encoding = tokenizer(
        test_ohg,
        return_tensors='pt',
        max_length=128,
        padding='max_length',
        truncation=True
    ).to(device)

    # Compute embeddings
    with torch.no_grad():
        latin_embedding = test_model(latin_encoding['input_ids'], latin_encoding['attention_mask'])
        ohg_embedding = test_model(ohg_encoding['input_ids'], ohg_encoding['attention_mask'])

    # Compute similarity
    similarity = torch.nn.functional.cosine_similarity(latin_embedding, ohg_embedding).item()
    print(f"\nLatin: {test_latin}")
    print(f"OHG: {test_ohg}")
    print(f"Similarity: {similarity:.4f}")

    return model

# PART 8: RUN THE TRAINING

# Start model training
print("Starting model training...")
trained_model = train_contrastive_model()

print("Training completed!")
```

```python
import os
import torch
import pandas as pd
import numpy as np
import torch.nn as nn
from torch.utils.data import Dataset
from transformers import AutoModel, AutoTokenizer
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
from google.colab import drive
from sklearn.metrics import confusion_matrix, classification_report

# Mount Google Drive
drive.mount('/content/drive')

# Set up paths
BASE_PATH = '/content/drive/MyDrive/latin_model'
MODEL_DIR = f'{BASE_PATH}/contrastive_model'
DATA_DIR = '/content/drive/MyDrive/OHG-texts'  # Update this to where your evaluation files are
RESULTS_DIR = f'{BASE_PATH}/evaluation_results'

# Create results directory if it doesn't exist
os.makedirs(RESULTS_DIR, exist_ok=True)

# Device configuration
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")

# Define the contrastive model (same as training)
class ContrastiveModel(nn.Module):
    def __init__(self, base_model_name='bert-base-multilingual-cased'):
        super(ContrastiveModel, self).__init__()
        self.bert = AutoModel.from_pretrained(base_model_name)

        # Projection layers
        self.projection = nn.Sequential(
            nn.Linear(768, 768),
            nn.ReLU(),
            nn.Linear(768, 128)
        )

    def forward(self, input_ids, attention_mask):
        outputs = self.bert(input_ids=input_ids, attention_mask=attention_mask)
        cls_output = outputs.last_hidden_state[:, 0, :]
        projection = self.projection(cls_output)
        projection = nn.functional.normalize(projection, p=2, dim=1)
        return projection

# Load the trained model
def load_model(model_path):
    print(f"Loading model from {model_path}...")
    model = ContrastiveModel().to(device)

    try:
        model.load_state_dict(torch.load(model_path, map_location=device))
        print("Model loaded successfully")
        return model
    except Exception as e:
        print(f"Error loading model: {e}")
        return None

# Function to compute similarity between two texts
def compute_similarity(model, tokenizer, latin_text, ohg_text, max_length=128):
    model.eval()

    # Tokenize the texts
    latin_encoding = tokenizer(
        latin_text,
        return_tensors='pt',
        max_length=max_length,
        padding='max_length',
        truncation=True
    ).to(device)

    ohg_encoding = tokenizer(
        ohg_text,
        return_tensors='pt',
        max_length=max_length,
        padding='max_length',
        truncation=True
    ).to(device)

    # Compute embeddings
    with torch.no_grad():
        latin_embedding = model(latin_encoding['input_ids'], latin_encoding['attention_mask'])
        ohg_embedding = model(ohg_encoding['input_ids'], ohg_encoding['attention_mask'])

    # Compute similarity
    similarity = torch.nn.functional.cosine_similarity(latin_embedding, ohg_embedding).item()
    return similarity

# Function to load evaluation data from Excel files
def load_evaluation_data(excel_path, source_name, latin_col='Source clause', ohg_col='Target clause', label_col=None):
    """
    Load evaluation data from Excel file
    If label_col is provided, it should contain the ground truth labels:
    'translation', 'paraphrase', 'adaptation', or 'unrelated'
    """
    try:
        print(f"Loading {source_name} from {excel_path}...")

        # Read the Excel file
        df = pd.read_excel(excel_path)

        print(f"  Using columns: '{latin_col}' (Latin) and '{ohg_col}' (OHG)")

        # Extract text pairs, skipping empty or null rows
        pairs = []

        for _, row in df.iterrows():
            latin = str(row[latin_col]).strip()
            ohg = str(row[ohg_col]).strip()

            # Skip rows where either text is empty, NaN, or \N
            if (latin and ohg and
                latin != 'nan' and ohg != 'nan' and
                latin != '\\N' and ohg != '\\N'):

                # If label column is provided, include the label
                if label_col and label_col in row:
                    label = row[label_col]
                    pairs.append((latin, ohg, label))
                else:
                    pairs.append((latin, ohg, None))

        print(f"  Extracted {len(pairs)} valid pairs from {source_name}")
        return pairs

    except Exception as e:
        print(f"Error processing {source_name}: {e}")
        return []

# Function to evaluate text pairs
def evaluate_pairs(model, tokenizer, pairs, thresholds=None):
    """
    Evaluate similarity for a list of text pairs

    Args:
        model: Trained contrastive model
        tokenizer: Tokenizer for the model
        pairs: List of tuples (latin_text, ohg_text, label)
        thresholds: Dictionary with thresholds for classification
            e.g., {'translation': 0.75, 'paraphrase': 0.65, 'adaptation': 0.5}

    Returns:
        DataFrame with results
    """
    if thresholds is None:
        thresholds = {
            'translation': 0.75,
            'paraphrase': 0.65,
            'adaptation': 0.5
        }

    results = []

    for latin, ohg, true_label in tqdm(pairs, desc="Evaluating pairs"):
        similarity = compute_similarity(model, tokenizer, latin, ohg)

        # Determine predicted label based on thresholds
        pred_label = 'unrelated'
        if similarity >= thresholds['translation']:
            pred_label = 'translation'
        elif similarity >= thresholds['paraphrase']:
            pred_label = 'paraphrase'
        elif similarity >= thresholds['adaptation']:
            pred_label = 'adaptation'

        results.append({
            'latin': latin,
            'ohg': ohg,
            'true_label': true_label,
            'similarity': similarity,
            'predicted_label': pred_label
        })

    return pd.DataFrame(results)

# Function to optimize thresholds based on evaluation results
def optimize_thresholds(results_df, threshold_range=(0.1, 0.9), step=0.05):
    """
    Optimize thresholds based on evaluation results

    Args:
        results_df: DataFrame with evaluation results (must include 'true_label' and 'similarity')
        threshold_range: Tuple with min and max threshold values to test
        step: Step size for threshold testing

    Returns:
        Dictionary with optimized thresholds
    """
    # Filter for rows with true labels
    labeled_df = results_df[results_df['true_label'].notna()].copy()

    if len(labeled_df) == 0:
        print("No labeled data available for threshold optimization")
        return {
            'translation': 0.75,
            'paraphrase': 0.65,
            'adaptation': 0.5
        }

    best_accuracy = 0
    best_thresholds = None

    # Create grid of thresholds to test
    thresholds = np.arange(threshold_range[0], threshold_range[1], step)

    for t_translation in thresholds:
        if t_translation < 0.5:  # Translation threshold should be higher
            continue

        for t_paraphrase in thresholds:
            if t_paraphrase >= t_translation or t_paraphrase < 0.4:
                continue

            for t_adaptation in thresholds:
                if t_adaptation >= t_paraphrase:
                    continue

                # Apply thresholds
                def get_label(similarity):
                    if similarity >= t_translation:
                        return 'translation'
                    elif similarity >= t_paraphrase:
                        return 'paraphrase'
                    elif similarity >= t_adaptation:
                        return 'adaptation'
                    else:
                        return 'unrelated'

                labeled_df['pred_label'] = labeled_df['similarity'].apply(get_label)

                # Calculate accuracy
                accuracy = (labeled_df['pred_label'] == labeled_df['true_label']).mean()

                if accuracy > best_accuracy:
                    best_accuracy = accuracy
                    best_thresholds = {
                        'translation': t_translation,
                        'paraphrase': t_paraphrase,
                        'adaptation': t_adaptation
                    }

    print(f"Best thresholds found with accuracy {best_accuracy:.4f}:")
    print(f"  Translation: >= {best_thresholds['translation']:.2f}")
    print(f"  Paraphrase:  >= {best_thresholds['paraphrase']:.2f}")
    print(f"  Adaptation:  >= {best_thresholds['adaptation']:.2f}")
    print(f"  Unrelated:   < {best_thresholds['adaptation']:.2f}")

    return best_thresholds

# Function to visualize similarity distribution
def visualize_similarity_distribution(results_df, output_path=None):
    """Visualize distribution of similarity scores by label"""
    plt.figure(figsize=(12, 6))

    if 'true_label' in results_df and results_df['true_label'].notna().any():
        # Use true labels if available
        sns.histplot(data=results_df, x='similarity', hue='true_label', bins=30, alpha=0.7)
        plt.title('Distribution of Similarity Scores by True Label')
    else:
        # Otherwise just use predicted labels
        sns.histplot(data=results_df, x='similarity', hue='predicted_label', bins=30, alpha=0.7)
        plt.title('Distribution of Similarity Scores by Predicted Label')

    plt.xlabel('Cosine Similarity')
    plt.ylabel('Count')
    plt.grid(True, alpha=0.3)

    if output_path:
        plt.savefig(output_path)
        print(f"Saved similarity distribution plot to {output_path}")

    plt.show()

# Function to evaluate model performance with confusion matrix
def evaluate_model_performance(results_df, thresholds=None):
    """Evaluate model performance if true labels are available"""
    labeled_df = results_df[results_df['true_label'].notna()].copy()

    if len(labeled_df) == 0:
        print("No labeled data available for evaluation")
        return

    if thresholds:
        # Apply thresholds to update predicted labels
        def get_label(similarity):
            if similarity >= thresholds['translation']:
                return 'translation'
            elif similarity >= thresholds['paraphrase']:
                return 'paraphrase'
            elif similarity >= thresholds['adaptation']:
                return 'adaptation'
            else:
                return 'unrelated'

        labeled_df['predicted_label'] = labeled_df['similarity'].apply(get_label)

    # Get unique labels in order of "quality"
    all_labels = ['translation', 'paraphrase', 'adaptation', 'unrelated']
    labels = [label for label in all_labels if label in labeled_df['true_label'].unique()]

    # Compute confusion matrix
    cm = confusion_matrix(
        labeled_df['true_label'],
        labeled_df['predicted_label'],
        labels=labels
    )

    # Plot confusion matrix
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=labels, yticklabels=labels)
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.title('Confusion Matrix')
    plt.tight_layout()
    plt.savefig(os.path.join(RESULTS_DIR, 'confusion_matrix.png'))
    plt.show()

    # Print classification report
    report = classification_report(labeled_df['true_label'], labeled_df['predicted_label'])
    print("\nClassification Report:")
    print(report)

# Main function for supervised evaluation
def run_supervised_evaluation():
    """Run supervised evaluation on test datasets"""
    # Load the model
    model_path = os.path.join(MODEL_DIR, 'best_contrastive_model.pt')
    model = load_model(model_path)

    if model is None:
        print("Could not load model. Please check the model path.")
        return

    # Load tokenizer
    tokenizer = AutoTokenizer.from_pretrained('bert-base-multilingual-cased')

    # Current thresholds (from initial training)
    initial_thresholds = {
        'translation': 0.75,
        'paraphrase': 0.65,
        'adaptation': 0.5
    }

    # Example of loading Benedictine Rule evaluation data
    # Update the file path and column names to match your data
    benedictine_path = os.path.join(DATA_DIR, 'Benedictine_Rule_aligned.xlsx')

    # Check if file exists
    if not os.path.exists(benedictine_path):
        print(f"Warning: {benedictine_path} does not exist.")
        print("Please provide the correct path to your evaluation data.")

        # Alternative: List available Excel files in the directory
        excel_files = [f for f in os.listdir(DATA_DIR) if f.endswith('.xlsx') or f.endswith('.xls')]
        if excel_files:
            print("\nAvailable Excel files in the directory:")
            for idx, file in enumerate(excel_files):
                print(f"{idx+1}. {file}")

            # Let user select a file
            selection = input("Enter the number of the file you want to use for evaluation: ")
            try:
                idx = int(selection) - 1
                if 0 <= idx < len(excel_files):
                    benedictine_path = os.path.join(DATA_DIR, excel_files[idx])
                    print(f"Using {benedictine_path} for evaluation")
                else:
                    print("Invalid selection. Exiting.")
                    return
            except ValueError:
                print("Invalid input. Exiting.")
                return
        else:
            print("No Excel files found in the directory. Please specify the correct directory.")
            return

    # Load evaluation data
    # If your data has ground truth labels, specify the label column name
    # For example: label_col='Relation_Type'
    evaluation_pairs = load_evaluation_data(
        benedictine_path,
        'Benedictine Rule',
        latin_col='Source clause',  # Update these column names as needed
        ohg_col='Target clause',
        label_col=None  # Set to the name of your label column if available
    )

    if not evaluation_pairs:
        print("No evaluation pairs were loaded. Please check your data file.")
        return

    # Evaluate pairs
    results = evaluate_pairs(model, tokenizer, evaluation_pairs, thresholds=initial_thresholds)

    # Save results to CSV
    results_path = os.path.join(RESULTS_DIR, 'evaluation_results.csv')
    results.to_csv(results_path, index=False)
    print(f"Saved evaluation results to {results_path}")

    # Visualize similarity distribution
    plot_path = os.path.join(RESULTS_DIR, 'similarity_distribution.png')
    visualize_similarity_distribution(results, plot_path)

    # Optimize thresholds if true labels are available
    if 'true_label' in results and results['true_label'].notna().any():
        optimized_thresholds = optimize_thresholds(results)

        # Re-evaluate with optimized thresholds
        updated_results = evaluate_pairs(model, tokenizer, evaluation_pairs, thresholds=optimized_thresholds)

        # Save updated results
        updated_results_path = os.path.join(RESULTS_DIR, 'evaluation_results_optimized.csv')
        updated_results.to_csv(updated_results_path, index=False)
        print(f"Saved optimized evaluation results to {updated_results_path}")

        # Evaluate model performance
        evaluate_model_performance(updated_results, optimized_thresholds)
    else:
        print("\nNo ground truth labels available for threshold optimization.")
        print("Using initial thresholds for evaluation:")
        print(f"  Translation: >= {initial_thresholds['translation']:.2f}")
        print(f"  Paraphrase:  >= {initial_thresholds['paraphrase']:.2f}")
        print(f"  Adaptation:  >= {initial_thresholds['adaptation']:.2f}")
        print(f"  Unrelated:   < {initial_thresholds['adaptation']:.2f}")

    # Return results for further analysis
    return results

# Function to interactively test the model
def interactive_test():
    """Interactive testing of the model with user input"""
    # Load the model
    model_path = os.path.join(MODEL_DIR, 'best_contrastive_model.pt')
    model = load_model(model_path)

    if model is None:
        print("Could not load model. Please check the model path.")
        return

    # Load tokenizer
    tokenizer = AutoTokenizer.from_pretrained('bert-base-multilingual-cased')

    print("\n=== Interactive Testing Mode ===")
    print("Enter Latin and OHG text pairs to compute similarity")
    print("Enter 'q' to quit")

    while True:
        print("\n")
        latin = input("Enter Latin text: ")
        if latin.lower() == 'q':
            break

        ohg = input("Enter OHG text: ")
        if ohg.lower() == 'q':
            break

        similarity = compute_similarity(model, tokenizer, latin, ohg)

        # Determine relationship based on thresholds
        relationship = "unrelated"
        if similarity >= 0.75:
            relationship = "translation"
        elif similarity >= 0.65:
            relationship = "paraphrase"
        elif similarity >= 0.5:
            relationship = "adaptation"

        print(f"Similarity: {similarity:.4f}")
        print(f"Relationship: {relationship}")

# Run the evaluation
if __name__ == "__main__":
    # Run supervised evaluation
    results = run_supervised_evaluation()

    # Ask if user wants to test interactively
    test_interactively = input("\nWould you like to test the model interactively? (y/n): ")
    if test_interactively.lower() == 'y':
        interactive_test()

    print("\nEvaluation completed!")
```

```python
import os
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from transformers import AutoModel, AutoTokenizer
import numpy as np
from tqdm import tqdm
import pandas as pd
import matplotlib.pyplot as plt
import random
from google.colab import drive
from torch.cuda.amp import autocast, GradScaler

# Mount Google Drive
drive.mount('/content/drive')

# Set up paths
BASE_PATH = '/content/drive/MyDrive/latin_model'
MODEL_DIR = f'{BASE_PATH}/contrastive_model_improved'  # New directory for improved model
DATA_DIR = '/content/drive/MyDrive/OHG-texts'

# Create model directory if it doesn't exist
os.makedirs(MODEL_DIR, exist_ok=True)

print(f"Improved model will be saved to: {MODEL_DIR}")

# Device configuration
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")

# Define the contrastive model (same architecture as before)
class ContrastiveModel(nn.Module):
    def __init__(self, base_model_name='bert-base-multilingual-cased'):
        super(ContrastiveModel, self).__init__()
        self.bert = AutoModel.from_pretrained(base_model_name)

        # Projection layers
        self.projection = nn.Sequential(
            nn.Linear(768, 768),
            nn.ReLU(),
            nn.Linear(768, 128)
        )

    def forward(self, input_ids, attention_mask):
        outputs = self.bert(input_ids=input_ids, attention_mask=attention_mask)
        cls_output = outputs.last_hidden_state[:, 0, :]
        projection = self.projection(cls_output)
        projection = nn.functional.normalize(projection, p=2, dim=1)
        return projection

# Dataset class for parallel texts with various types of negatives
class EnhancedParallelTextDataset(Dataset):
    def __init__(self, pairs, tokenizer, max_length=128):
        """
        Initialize dataset with positive and negative pairs

        Args:
            pairs: List of tuples (latin_text, ohg_text, label)
                  where label is 1 for positive pairs (translations)
                  and 0 for negative pairs (non-translations)
            tokenizer: Tokenizer for encoding texts
            max_length: Maximum sequence length
        """
        self.pairs = pairs
        self.tokenizer = tokenizer
        self.max_length = max_length

    def __len__(self):
        return len(self.pairs)

    def __getitem__(self, idx):
        latin, ohg, label = self.pairs[idx]

        latin_encoding = self.tokenizer(
            latin,
            return_tensors='pt',
            max_length=self.max_length,
            padding='max_length',
            truncation=True
        )

        ohg_encoding = self.tokenizer(
            ohg,
            return_tensors='pt',
            max_length=self.max_length,
            padding='max_length',
            truncation=True
        )

        return {
            'latin_input_ids': latin_encoding['input_ids'].squeeze(),
            'latin_attention_mask': latin_encoding['attention_mask'].squeeze(),
            'ohg_input_ids': ohg_encoding['input_ids'].squeeze(),
            'ohg_attention_mask': ohg_encoding['attention_mask'].squeeze(),
            'label': torch.tensor(label, dtype=torch.float)
        }

# Enhanced contrastive loss with margin
class EnhancedContrastiveLoss(nn.Module):
    def __init__(self, temperature=0.5, margin=0.2):
        super(EnhancedContrastiveLoss, self).__init__()
        self.temperature = temperature
        self.margin = margin

    def forward(self, z_i, z_j, labels):
        """
        Compute contrastive loss with margin

        Args:
            z_i: Latin embeddings
            z_j: OHG embeddings
            labels: 1 for positive pairs, 0 for negative pairs
        """
        batch_size = z_i.size(0)

        # Compute similarity matrix
        sim = torch.mm(z_i, z_j.t()) / self.temperature

        # Diagonal elements are similarity between corresponding pairs
        diag_sim = torch.diag(sim)

        # For positive pairs: maximize similarity
        pos_loss = -torch.mean(diag_sim * labels)

        # For negative pairs: minimize similarity with margin
        # We want sim <= margin for negative pairs
        neg_loss = torch.mean(torch.relu(diag_sim - self.margin) * (1 - labels))

        return pos_loss + neg_loss

# Load model and tokenizer
def load_base_model():
    """Load the previously trained model as starting point"""
    base_model_path = os.path.join(BASE_PATH, 'contrastive_model/best_contrastive_model.pt')

    print(f"Loading base model from {base_model_path}...")
    model = ContrastiveModel().to(device)

    try:
        model.load_state_dict(torch.load(base_model_path, map_location=device))
        print("Base model loaded successfully!")
    except Exception as e:
        print(f"Error loading base model: {e}")
        print("Initializing a new model...")

    return model

# Save model function with verification
def save_model_robust(model, path):
    """Save model with verification"""
    print(f"Saving model to {path}...")
    torch.save(model.state_dict(), path)

    if os.path.exists(path):
        size_mb = os.path.getsize(path) / (1024 * 1024)
        print(f"✓ Model saved: {path} ({size_mb:.2f} MB)")

        # Force sync with Google Drive
        try:
            print("Syncing with Google Drive...")
            drive.flush_and_unmount()
            drive.mount('/content/drive')
            print("Google Drive sync complete")

            # Verify file still exists
            if os.path.exists(path):
                print(f"✓ Verified after sync: {path}")
                return True
            else:
                print(f"✗ File missing after sync: {path}")
                return False
        except Exception as e:
            print(f"Drive sync warning: {e}")
    else:
        print(f"✗ Failed to save model: {path}")
        return False

# Function to load data from Excel files
def load_excel_aligned_texts(excel_path, source_name):
    """Extract parallel texts from Excel files with aligned Latin and OHG"""
    try:
        print(f"Loading {source_name} from {excel_path}...")

        # Read the Excel file
        df = pd.read_excel(excel_path)

        # Use specific column names
        latin_col = 'Source clause'
        ohg_col = 'Target clause'

        print(f"  Using columns: '{latin_col}' (Latin) and '{ohg_col}' (OHG)")

        # Extract text pairs, skipping empty or null rows
        latin_texts = []
        ohg_texts = []

        for _, row in df.iterrows():
            latin = str(row[latin_col]).strip()
            ohg = str(row[ohg_col]).strip()

            # Skip rows where either text is empty, NaN, or \N
            if (latin and ohg and
                latin != 'nan' and ohg != 'nan' and
                latin != '\\N' and ohg != '\\N'):
                latin_texts.append(latin)
                ohg_texts.append(ohg)

        print(f"  Extracted {len(latin_texts)} valid pairs from {source_name}")
        return latin_texts, ohg_texts

    except Exception as e:
        print(f"Error processing {source_name}: {e}")
        return [], []

# Function to create different types of negative pairs
def create_negative_pairs(latin_texts, ohg_texts):
    """
    Create various types of negative pairs:
    1. Random negatives: Random Latin-OHG pairs
    2. Hard negatives: Sentences with similar structure but different meaning
    3. Semi-hard negatives: Sentences from same text but not translations
    """
    all_negative_pairs = []
    n = len(latin_texts)

    # 1. Random negatives (completely different texts)
    print("Creating random negative pairs...")
    random_negative_count = n // 2  # 50% of positive pairs
    for _ in range(random_negative_count):
        idx1 = random.randint(0, n-1)
        idx2 = random.randint(0, n-1)

        # Ensure the indices are different (not the same pair)
        while idx2 == idx1:
            idx2 = random.randint(0, n-1)

        all_negative_pairs.append((latin_texts[idx1], ohg_texts[idx2], 0))

    # 2. Hard negatives (sentences with similar length/structure)
    print("Creating hard negative pairs based on structural similarity...")

    # Group sentences by length as a simple proxy for structure
    length_groups_latin = {}
    for i, text in enumerate(latin_texts):
        length = len(text.split())
        if length not in length_groups_latin:
            length_groups_latin[length] = []
        length_groups_latin[length].append(i)

    length_groups_ohg = {}
    for i, text in enumerate(ohg_texts):
        length = len(text.split())
        if length not in length_groups_ohg:
            length_groups_ohg[length] = []
        length_groups_ohg[length].append(i)

    hard_negative_count = n // 4  # 25% of positive pairs
    hard_negatives_added = 0

    # Try to find sentences with similar lengths
    for length in length_groups_latin:
        if hard_negatives_added >= hard_negative_count:
            break

        if length in length_groups_ohg and len(length_groups_latin[length]) > 0 and len(length_groups_ohg[length]) > 0:
            # Get all possible pairs within this length group
            latin_indices = length_groups_latin[length]
            ohg_indices = length_groups_ohg[length]

            for latin_idx in latin_indices:
                if hard_negatives_added >= hard_negative_count:
                    break

                for ohg_idx in ohg_indices:
                    # Ensure this isn't a positive pair (assuming aligned indices are translations)
                    if latin_idx != ohg_idx:
                        all_negative_pairs.append((latin_texts[latin_idx], ohg_texts[ohg_idx], 0))
                        hard_negatives_added += 1

                        if hard_negatives_added >= hard_negative_count:
                            break

    # 3. Semi-hard negatives (from same text but not translations)
    print("Creating semi-hard negative pairs from same texts...")
    semi_hard_negative_count = n // 4  # 25% of positive pairs

    # Create pairs that are close in the text but not direct translations
    for i in range(n - 2):
        if len(all_negative_pairs) >= random_negative_count + hard_negative_count + semi_hard_negative_count:
            break

        # Use Latin i with OHG i+2 (skipping direct neighbors)
        all_negative_pairs.append((latin_texts[i], ohg_texts[min(i+2, n-1)], 0))

    print(f"Created {len(all_negative_pairs)} negative pairs")
    return all_negative_pairs

# Function to train the model with improved contrastive learning
def train_model_with_contrastive_examples(model, train_loader, val_loader, epochs=3, lr=1e-5):
    """Train the model with positive and negative examples"""
    optimizer = torch.optim.Adam(model.parameters(), lr=lr)
    criterion = EnhancedContrastiveLoss(temperature=0.5, margin=0.2)

    # For mixed precision training on GPU
    scaler = GradScaler() if device.type == 'cuda' else None

    train_losses = []
    val_losses = []
    best_val_loss = float('inf')

    for epoch in range(epochs):
        # Training
        model.train()
        epoch_loss = 0
        batch_count = 0

        progress_bar = tqdm(train_loader, desc=f"Epoch {epoch+1}/{epochs}")
        for batch in progress_bar:
            latin_input_ids = batch['latin_input_ids'].to(device)
            latin_attention_mask = batch['latin_attention_mask'].to(device)
            ohg_input_ids = batch['ohg_input_ids'].to(device)
            ohg_attention_mask = batch['ohg_attention_mask'].to(device)
            labels = batch['label'].to(device)

            # Using mixed precision for GPU training
            if device.type == 'cuda':
                with autocast():
                    # Forward pass
                    latin_embeddings = model(latin_input_ids, latin_attention_mask)
                    ohg_embeddings = model(ohg_input_ids, ohg_attention_mask)

                    # Compute loss
                    loss = criterion(latin_embeddings, ohg_embeddings, labels)

                # Backward and optimize with scaling
                optimizer.zero_grad()
                scaler.scale(loss).backward()
                scaler.step(optimizer)
                scaler.update()
            else:
                # Forward pass
                latin_embeddings = model(latin_input_ids, latin_attention_mask)
                ohg_embeddings = model(ohg_input_ids, ohg_attention_mask)

                # Compute loss
                loss = criterion(latin_embeddings, ohg_embeddings, labels)

                # Backward and optimize
                optimizer.zero_grad()
                loss.backward()
                optimizer.step()

            epoch_loss += loss.item()
            batch_count += 1

            progress_bar.set_postfix({'loss': loss.item()})

        avg_train_loss = epoch_loss / batch_count
        train_losses.append(avg_train_loss)

        # Validation
        model.eval()
        val_loss = 0
        val_batch_count = 0

        with torch.no_grad():
            for batch in val_loader:
                latin_input_ids = batch['latin_input_ids'].to(device)
                latin_attention_mask = batch['latin_attention_mask'].to(device)
                ohg_input_ids = batch['ohg_input_ids'].to(device)
                ohg_attention_mask = batch['ohg_attention_mask'].to(device)
                labels = batch['label'].to(device)

                # Forward pass
                latin_embeddings = model(latin_input_ids, latin_attention_mask)
                ohg_embeddings = model(ohg_input_ids, ohg_attention_mask)

                # Compute loss
                loss = criterion(latin_embeddings, ohg_embeddings, labels)
                val_loss += loss.item()
                val_batch_count += 1

        avg_val_loss = val_loss / val_batch_count
        val_losses.append(avg_val_loss)

        print(f"Epoch {epoch+1}/{epochs}, Train loss: {avg_train_loss:.4f}, Val loss: {avg_val_loss:.4f}")

        # Save best model
        if avg_val_loss < best_val_loss:
            best_val_loss = avg_val_loss
            best_model_path = os.path.join(MODEL_DIR, 'best_contrastive_model_improved.pt')
            save_model_robust(model, best_model_path)

    # Save final model
    final_model_path = os.path.join(MODEL_DIR, 'final_contrastive_model_improved.pt')
    save_model_robust(model, final_model_path)

    # Plot training curve
    plt.figure(figsize=(10, 5))
    plt.plot(train_losses, label='Training Loss')
    plt.plot(val_losses, label='Validation Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.title('Training and Validation Loss')
    plt.legend()
    plt.savefig(os.path.join(MODEL_DIR, 'training_curve_improved.png'))

    return model, train_losses, val_losses

# Main function to run the improved training
def run_improved_training():
    # Load tokenizer
    tokenizer = AutoTokenizer.from_pretrained('bert-base-multilingual-cased')

    # Load base model
    model = load_base_model()

    # Load data from Excel files
    tatian_latin, tatian_ohg = load_excel_aligned_texts(
        os.path.join(DATA_DIR, 'Tatian_aligned.xlsx'), 'Tatian')

    isidor_latin, isidor_ohg = load_excel_aligned_texts(
        os.path.join(DATA_DIR, 'Isidor_aligned.xlsx'), 'Isidor')

    physio_latin, physio_ohg = load_excel_aligned_texts(
        os.path.join(DATA_DIR, 'Physiologus_aligned.xlsx'), 'Physiologus')

    benedict_latin, benedict_ohg = load_excel_aligned_texts(
        os.path.join(DATA_DIR, 'Benedictine_Rule_aligned.xlsx'), 'Benedictine Rule')

    # Combine all texts
    all_latin = tatian_latin + isidor_latin + physio_latin + benedict_latin
    all_ohg = tatian_ohg + isidor_ohg + physio_ohg + benedict_ohg

    print(f"Total loaded: {len(all_latin)} parallel texts")

    # Create positive pairs (true translations)
    positive_pairs = [(latin, ohg, 1) for latin, ohg in zip(all_latin, all_ohg)]

    # Create negative pairs (non-translations)
    negative_pairs = create_negative_pairs(all_latin, all_ohg)

    # Combine positive and negative pairs
    all_pairs = positive_pairs + negative_pairs

    # Shuffle the data
    random.shuffle(all_pairs)

    # Split into training and validation sets (90/10 split)
    split_idx = int(0.9 * len(all_pairs))
    train_pairs, val_pairs = all_pairs[:split_idx], all_pairs[split_idx:]

    print(f"Training set: {len(train_pairs)} pairs")
    print(f"Validation set: {len(val_pairs)} pairs")

    # Create datasets and dataloaders
    train_dataset = EnhancedParallelTextDataset(train_pairs, tokenizer)
    val_dataset = EnhancedParallelTextDataset(val_pairs, tokenizer)

    batch_size = 16
    if device.type == 'cuda':
        batch_size = 32  # Larger batch size for GPU

    train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
    val_loader = DataLoader(val_dataset, batch_size=batch_size)

    # Train the model
    improved_model, train_losses, val_losses = train_model_with_contrastive_examples(
        model,
        train_loader,
        val_loader,
        epochs=3
    )

    # Test the improved model
    test_improved_model(improved_model, tokenizer)

    return improved_model

# Function to test the improved model
def test_improved_model(model, tokenizer):
    """Test the improved model on examples"""
    model.eval()

    # Test cases
    test_cases = [
        # Positive case (true translation)
        {
            "latin": "Monachorum quattuor esse genera manifestum est.",
            "ohg": "municho fioreo uuesan chunni kund ist.",
            "expected": "translation"
        },
        # Negative case (different content)
        {
            "latin": "Quartum vero genus est monachorum, quod nominatur gyrovagum.",
            "ohg": "Únz mír sâldâ fólgetôn . in állemo mînemo gûote.",
            "expected": "unrelated"
        }
    ]

    print("\nTesting improved model:")
    for test in test_cases:
        latin = test["latin"]
        ohg = test["ohg"]
        expected = test["expected"]

        # Tokenize
        latin_encoding = tokenizer(
            latin,
            return_tensors='pt',
            max_length=128,
            padding='max_length',
            truncation=True
        ).to(device)

        ohg_encoding = tokenizer(
            ohg,
            return_tensors='pt',
            max_length=128,
            padding='max_length',
            truncation=True
        ).to(device)

        # Compute embeddings
        with torch.no_grad():
            latin_embedding = model(latin_encoding['input_ids'], latin_encoding['attention_mask'])
            ohg_embedding = model(ohg_encoding['input_ids'], ohg_encoding['attention_mask'])

        # Compute similarity
        similarity = torch.nn.functional.cosine_similarity(latin_embedding, ohg_embedding).item()

        # Determine relationship based on similarity
        relationship = "unrelated"
        if similarity >= 0.75:
            relationship = "translation"
        elif similarity >= 0.65:
            relationship = "paraphrase"
        elif similarity >= 0.5:
            relationship = "adaptation"

        print(f"\nLatin: {latin}")
        print(f"OHG: {ohg}")
        print(f"Similarity: {similarity:.4f}")
        print(f"Predicted relationship: {relationship}")
        print(f"Expected relationship: {expected}")
        print(f"Match: {'✓' if relationship == expected else '✗'}")

# Run the improved training
if __name__ == "__main__":
    improved_model = run_improved_training()
    print("Improved training completed!")
```

```python
import os
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from transformers import AutoModel, AutoTokenizer
import numpy as np
from tqdm import tqdm
import pandas as pd
import matplotlib.pyplot as plt
import random
from google.colab import drive
import torch.nn.functional as F

# Mount Google Drive
drive.mount('/content/drive')

# Set up paths
BASE_PATH = '/content/drive/MyDrive/latin_model'
MODEL_DIR = f'{BASE_PATH}/contrastive_model_enhanced'  # New directory for enhanced model
DATA_DIR = '/content/drive/MyDrive/OHG-texts'

# Create model directory if it doesn't exist
os.makedirs(MODEL_DIR, exist_ok=True)

print(f"Enhanced model will be saved to: {MODEL_DIR}")

# Device configuration
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")

# Define the contrastive model with enhanced projection
class EnhancedContrastiveModel(nn.Module):
    def __init__(self, base_model_name='bert-base-multilingual-cased'):
        super(EnhancedContrastiveModel, self).__init__()
        self.bert = AutoModel.from_pretrained(base_model_name)

        # Enhanced projection layers - deeper network for better semantic understanding
        self.projection = nn.Sequential(
            nn.Linear(768, 768),
            nn.LayerNorm(768),
            nn.ReLU(),
            nn.Dropout(0.1),
            nn.Linear(768, 384),
            nn.LayerNorm(384),
            nn.ReLU(),
            nn.Dropout(0.1),
            nn.Linear(384, 128)
        )

    def forward(self, input_ids, attention_mask):
        outputs = self.bert(input_ids=input_ids, attention_mask=attention_mask)
        cls_output = outputs.last_hidden_state[:, 0, :]
        projection = self.projection(cls_output)
        projection = F.normalize(projection, p=2, dim=1)
        return projection

# Dataset class with stronger negative sampling
class EnhancedContrastiveDataset(Dataset):
    def __init__(self, positive_pairs, negative_pairs, tokenizer, max_length=128):
        """
        Initialize dataset with explicit positive and negative pairs

        Args:
            positive_pairs: List of tuples (latin_text, ohg_text)
            negative_pairs: List of tuples (latin_text, ohg_text)
            tokenizer: Tokenizer for encoding texts
            max_length: Maximum sequence length
        """
        self.positive_pairs = positive_pairs
        self.negative_pairs = negative_pairs
        self.tokenizer = tokenizer
        self.max_length = max_length

    def __len__(self):
        return len(self.positive_pairs) + len(self.negative_pairs)

    def __getitem__(self, idx):
        # Determine if this is a positive or negative example
        if idx < len(self.positive_pairs):
            latin, ohg = self.positive_pairs[idx]
            label = 1.0  # Positive pair
        else:
            # Adjust index for negative pairs
            adj_idx = idx - len(self.positive_pairs)
            latin, ohg = self.negative_pairs[adj_idx]
            label = 0.0  # Negative pair

        latin_encoding = self.tokenizer(
            latin,
            return_tensors='pt',
            max_length=self.max_length,
            padding='max_length',
            truncation=True
        )

        ohg_encoding = self.tokenizer(
            ohg,
            return_tensors='pt',
            max_length=self.max_length,
            padding='max_length',
            truncation=True
        )

        return {
            'latin_input_ids': latin_encoding['input_ids'].squeeze(),
            'latin_attention_mask': latin_encoding['attention_mask'].squeeze(),
            'ohg_input_ids': ohg_encoding['input_ids'].squeeze(),
            'ohg_attention_mask': ohg_encoding['attention_mask'].squeeze(),
            'label': torch.tensor(label, dtype=torch.float)
        }

# Enhanced contrastive loss with higher margin
class StrongContrastiveLoss(nn.Module):
    def __init__(self, margin=0.8, pos_weight=1.0, neg_weight=2.0):
        """
        Enhanced contrastive loss with higher margin and weighting

        Args:
            margin: Minimum distance between positive and negative pairs (higher = stronger separation)
            pos_weight: Weight for positive pairs
            neg_weight: Weight for negative pairs (higher = stronger penalty)
        """
        super(StrongContrastiveLoss, self).__init__()
        self.margin = margin
        self.pos_weight = pos_weight
        self.neg_weight = neg_weight

    def forward(self, anchor_emb, positive_emb, negative_emb=None, labels=None):
        """
        Compute triplet-style contrastive loss with margin

        Args:
            anchor_emb: Latin embeddings
            positive_emb: OHG embeddings for positive pairs (translations)
            negative_emb: OHG embeddings for negative pairs (non-translations)
            labels: 1 for positive pairs, 0 for negative pairs
        """
        if labels is not None:
            # When using single batch with labels
            batch_size = anchor_emb.size(0)

            # Calculate cosine similarity
            similarity = torch.matmul(anchor_emb, positive_emb.t())

            # Extract diagonal elements (similarity between corresponding pairs)
            pos_similarity = torch.diag(similarity)

            # Apply weights based on labels
            pos_mask = labels > 0.5
            neg_mask = ~pos_mask

            # Loss for positive pairs: should have high similarity
            pos_loss = -torch.mean(pos_similarity[pos_mask]) * self.pos_weight

            # Loss for negative pairs: should have low similarity
            neg_loss = torch.mean(torch.relu(pos_similarity[neg_mask] - (1 - self.margin))) * self.neg_weight

            return pos_loss + neg_loss
        else:
            # Triplet-style with separate negative embeddings
            # Positive similarity (should be high)
            pos_similarity = F.cosine_similarity(anchor_emb, positive_emb)

            # Negative similarity (should be low)
            neg_similarity = F.cosine_similarity(anchor_emb, negative_emb)

            # Loss: push positive pairs together, negative pairs apart
            loss = torch.relu(self.margin - pos_similarity + neg_similarity)

            return loss.mean()

# Function to load the previously trained model
def load_base_model(target_model):
    """Load the previously trained model as starting point"""
    base_model_path = os.path.join(BASE_PATH, 'contrastive_model/best_contrastive_model.pt')

    print(f"Loading base model from {base_model_path}...")

    try:
        # Load state dict from previous model
        state_dict = torch.load(base_model_path, map_location=device)

        # Filter out incompatible keys (due to enhanced architecture)
        compatible_state_dict = {}
        for k, v in state_dict.items():
            if k.startswith('bert.'):
                compatible_state_dict[k] = v
            elif k.startswith('projection.0'):
                compatible_state_dict[k] = v

        # Load compatible weights
        target_model.load_state_dict(compatible_state_dict, strict=False)
        print("Base model BERT layers loaded successfully!")
    except Exception as e:
        print(f"Error loading base model: {e}")
        print("Initializing a new model...")

    return target_model

# Save model function with verification
def save_model_robust(model, path):
    """Save model with verification"""
    print(f"Saving model to {path}...")
    torch.save(model.state_dict(), path)

    if os.path.exists(path):
        size_mb = os.path.getsize(path) / (1024 * 1024)
        print(f"✓ Model saved: {path} ({size_mb:.2f} MB)")

        # Force sync with Google Drive
        try:
            print("Syncing with Google Drive...")
            drive.flush_and_unmount()
            drive.mount('/content/drive')
            print("Google Drive sync complete")

            # Verify file still exists
            if os.path.exists(path):
                print(f"✓ Verified after sync: {path}")
                return True
            else:
                print(f"✗ File missing after sync: {path}")
                return False
        except Exception as e:
            print(f"Drive sync warning: {e}")
    else:
        print(f"✗ Failed to save model: {path}")
        return False

# Function to load data from Excel files
def load_excel_aligned_texts(excel_path, source_name):
    """Extract parallel texts from Excel files with aligned Latin and OHG"""
    try:
        print(f"Loading {source_name} from {excel_path}...")

        # Read the Excel file
        df = pd.read_excel(excel_path)

        # Use specific column names
        latin_col = 'Source clause'
        ohg_col = 'Target clause'

        print(f"  Using columns: '{latin_col}' (Latin) and '{ohg_col}' (OHG)")

        # Extract text pairs, skipping empty or null rows
        latin_texts = []
        ohg_texts = []

        for _, row in df.iterrows():
            latin = str(row[latin_col]).strip()
            ohg = str(row[ohg_col]).strip()

            # Skip rows where either text is empty, NaN, or \N
            if (latin and ohg and
                latin != 'nan' and ohg != 'nan' and
                latin != '\\N' and ohg != '\\N'):
                latin_texts.append(latin)
                ohg_texts.append(ohg)

        print(f"  Extracted {len(latin_texts)} valid pairs from {source_name}")
        return latin_texts, ohg_texts

    except Exception as e:
        print(f"Error processing {source_name}: {e}")
        return [], []

# Function to generate stronger negative pairs
def create_stronger_negative_pairs(latin_texts, ohg_texts):
    """
    Create more challenging negative pairs for better contrast
    """
    print("Creating stronger negative pairs...")
    all_negative_pairs = []
    n = len(latin_texts)

    # 1. Random negatives (approximately equal to positive pairs)
    random_negative_count = n

    # Track which pairs we've already made to avoid duplicates
    used_pairs = set()
    for _ in range(random_negative_count):
        idx1 = random.randint(0, n-1)
        idx2 = random.randint(0, n-1)

        # Ensure this isn't a positive pair (aligned indices)
        # and that we haven't already created this negative pair
        pair_key = (idx1, idx2)
        while idx2 == idx1 or pair_key in used_pairs:
            idx2 = random.randint(0, n-1)
            pair_key = (idx1, idx2)

        used_pairs.add(pair_key)
        all_negative_pairs.append((latin_texts[idx1], ohg_texts[idx2]))

    # 2. Structure-preserving negatives (based on length similarity)
    # Group texts by length
    length_buckets = {}
    for i, latin in enumerate(latin_texts):
        length = len(latin.split())
        if length not in length_buckets:
            length_buckets[length] = []
        length_buckets[length].append(i)

    # For each Latin text, find length-similar OHG text that's not its translation
    structure_negative_count = min(n, 1000)  # Cap at 1000 examples
    structure_negatives_added = 0

    for latin_idx in range(n):
        if structure_negatives_added >= structure_negative_count:
            break

        latin_length = len(latin_texts[latin_idx].split())

        # Look for length matches within ±1 word
        potential_matches = []
        for length in range(latin_length - 1, latin_length + 2):
            if length in length_buckets:
                potential_matches.extend(length_buckets[length])

        # Filter out the aligned index and previously used pairs
        potential_matches = [idx for idx in potential_matches
                            if idx != latin_idx and (latin_idx, idx) not in used_pairs]

        if potential_matches:
            ohg_idx = random.choice(potential_matches)
            used_pairs.add((latin_idx, ohg_idx))
            all_negative_pairs.append((latin_texts[latin_idx], ohg_texts[ohg_idx]))
            structure_negatives_added += 1

    # 3. Adversarial negatives (shuffle words within a sentence)
    adversarial_count = min(n // 2, 500)  # Cap at 500 examples

    for i in range(adversarial_count):
        idx = random.randint(0, n-1)

        # Take original OHG text and shuffle its words
        ohg_words = ohg_texts[idx].split()
        if len(ohg_words) > 3:  # Only shuffle if there are enough words
            random.shuffle(ohg_words)
            shuffled_ohg = ' '.join(ohg_words)

            # Add as negative pair
            all_negative_pairs.append((latin_texts[idx], shuffled_ohg))

    print(f"Created {len(all_negative_pairs)} negative pairs")
    return all_negative_pairs

# Enhanced training function
def train_with_strong_negatives(model, train_loader, val_loader, epochs=5, lr=2e-5):
    """Train with stronger emphasis on separating negative examples"""
    optimizer = torch.optim.AdamW(model.parameters(), lr=lr, weight_decay=0.01)

    # Use stronger contrastive loss
    criterion = StrongContrastiveLoss(margin=0.8, pos_weight=1.0, neg_weight=3.0)

    # Learning rate scheduler
    scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(
        optimizer, mode='min', factor=0.5, patience=1, verbose=True
    )

    train_losses = []
    val_losses = []
    best_val_loss = float('inf')

    for epoch in range(epochs):
        # Training
        model.train()
        epoch_loss = 0
        batch_count = 0

        progress_bar = tqdm(train_loader, desc=f"Epoch {epoch+1}/{epochs}")
        for batch in progress_bar:
            latin_input_ids = batch['latin_input_ids'].to(device)
            latin_attention_mask = batch['latin_attention_mask'].to(device)
            ohg_input_ids = batch['ohg_input_ids'].to(device)
            ohg_attention_mask = batch['ohg_attention_mask'].to(device)
            labels = batch['label'].to(device)

            # Forward pass
            latin_embeddings = model(latin_input_ids, latin_attention_mask)
            ohg_embeddings = model(ohg_input_ids, ohg_attention_mask)

            # Compute loss
            loss = criterion(latin_embeddings, ohg_embeddings, labels=labels)

            # Backward and optimize
            optimizer.zero_grad()
            loss.backward()

            # Gradient clipping to prevent exploding gradients
            torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)

            optimizer.step()

            epoch_loss += loss.item()
            batch_count += 1

            progress_bar.set_postfix({'loss': loss.item()})

        avg_train_loss = epoch_loss / batch_count
        train_losses.append(avg_train_loss)

        # Validation
        model.eval()
        val_loss = 0
        val_batch_count = 0

        with torch.no_grad():
            for batch in val_loader:
                latin_input_ids = batch['latin_input_ids'].to(device)
                latin_attention_mask = batch['latin_attention_mask'].to(device)
                ohg_input_ids = batch['ohg_input_ids'].to(device)
                ohg_attention_mask = batch['ohg_attention_mask'].to(device)
                labels = batch['label'].to(device)

                # Forward pass
                latin_embeddings = model(latin_input_ids, latin_attention_mask)
                ohg_embeddings = model(ohg_input_ids, ohg_attention_mask)

                # Compute loss
                loss = criterion(latin_embeddings, ohg_embeddings, labels=labels)
                val_loss += loss.item()
                val_batch_count += 1

        avg_val_loss = val_loss / val_batch_count
        val_losses.append(avg_val_loss)

        print(f"Epoch {epoch+1}/{epochs}, Train loss: {avg_train_loss:.4f}, Val loss: {avg_val_loss:.4f}")

        # Update learning rate based on validation loss
        scheduler.step(avg_val_loss)

        # Save best model
        if avg_val_loss < best_val_loss:
            best_val_loss = avg_val_loss
            best_model_path = os.path.join(MODEL_DIR, 'best_contrastive_model_enhanced.pt')
            save_model_robust(model, best_model_path)
            print(f"New best model saved! (Validation loss: {avg_val_loss:.4f})")

    # Save final model
    final_model_path = os.path.join(MODEL_DIR, 'final_contrastive_model_enhanced.pt')
    save_model_robust(model, final_model_path)

    # Plot training curve
    plt.figure(figsize=(10, 5))
    plt.plot(train_losses, label='Training Loss')
    plt.plot(val_losses, label='Validation Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.title('Training and Validation Loss')
    plt.legend()
    plt.savefig(os.path.join(MODEL_DIR, 'training_curve_enhanced.png'))

    return model, train_losses, val_losses

# Function to evaluate the enhanced model
def evaluate_enhanced_model(model, tokenizer):
    """Evaluate model on test cases, with detailed metrics"""
    model.eval()

    # Test cases covering a range of relationships
    test_cases = [
        # True translation
        {
            "latin": "Monachorum quattuor esse genera manifestum est.",
            "ohg": "municho fioreo uuesan chunni kund ist.",
            "expected": "translation",
            "relationship": "translation"
        },
        # Previously problematic case (unrelated but scored high)
        {
            "latin": "Quartum vero genus est monachorum, quod nominatur gyrovagum.",
            "ohg": "Únz mír sâldâ fólgetôn . in állemo mînemo gûote.",
            "expected": "unrelated",
            "relationship": "unrelated"
        },
        # Another translation from Benedictine Rule
        {
            "latin": "Pro lege eis est desideriorum voluptas.",
            "ohg": "Fora euu im ist kiridono uunnilust.",
            "expected": "translation",
            "relationship": "translation"
        },
        # Shuffled words (should be lower similarity)
        {
            "latin": "Pro lege eis est desideriorum voluptas.",
            "ohg": "ist kiridono im Fora uunnilust euu.",
            "expected": "unrelated",
            "relationship": "shuffled"
        }
    ]

    print("\nEvaluating Enhanced Model:")
    print("=" * 50)

    results = []

    for test in test_cases:
        latin = test["latin"]
        ohg = test["ohg"]
        expected = test["expected"]
        relationship = test["relationship"]

        # Tokenize
        latin_encoding = tokenizer(
            latin,
            return_tensors='pt',
            max_length=128,
            padding='max_length',
            truncation=True
        ).to(device)

        ohg_encoding = tokenizer(
            ohg,
            return_tensors='pt',
            max_length=128,
            padding='max_length',
            truncation=True
        ).to(device)

        # Compute embeddings
        with torch.no_grad():
            latin_embedding = model(latin_encoding['input_ids'], latin_encoding['attention_mask'])
            ohg_embedding = model(ohg_encoding['input_ids'], ohg_encoding['attention_mask'])

        # Compute similarity
        similarity = F.cosine_similarity(latin_embedding, ohg_embedding).item()

        # Determine relationship based on higher thresholds
        # Using higher thresholds to be more selective about translations
        predicted = "unrelated"
        if similarity >= 0.85:  # Higher threshold for translation
            predicted = "translation"
        elif similarity >= 0.70:
            predicted = "paraphrase"
        elif similarity >= 0.55:
            predicted = "adaptation"

        # Store result
        results.append({
            "latin": latin,
            "ohg": ohg,
            "similarity": similarity,
            "predicted": predicted,
            "expected": expected,
            "relationship": relationship,
            "match": predicted == expected
        })

        # Print result
        print(f"\nTest Case ({relationship}):")
        print(f"Latin: {latin}")
        print(f"OHG: {ohg}")
        print(f"Similarity: {similarity:.4f}")
        print(f"Predicted: {predicted}")
        print(f"Expected: {expected}")
        print(f"Match: {'✓' if predicted == expected else '✗'}")

    # Calculate overall accuracy
    accuracy = sum(1 for r in results if r["match"]) / len(results)
    print("\nOverall accuracy:", f"{accuracy:.2%}")

    # Plot similarity distribution
    plt.figure(figsize=(10, 6))

    relationship_types = ["translation", "unrelated", "shuffled"]
    colors = ["green", "red", "orange"]

    for rtype, color in zip(relationship_types, colors):
        similarities = [r["similarity"] for r in results if r["relationship"] == rtype]
        if similarities:
            plt.hist(similarities, alpha=0.7, label=rtype, color=color, bins=5)

    plt.axvline(x=0.85, color='black', linestyle='--', label='Translation threshold')
    plt.axvline(x=0.70, color='gray', linestyle='--', label='Paraphrase threshold')
    plt.axvline(x=0.55, color='lightgray', linestyle='--', label='Adaptation threshold')

    plt.xlabel('Similarity Score')
    plt.ylabel('Count')
    plt.title('Similarity Distribution by Relationship Type')
    plt.legend()
    plt.savefig(os.path.join(MODEL_DIR, 'similarity_distribution.png'))

    return results

# Main function to run the enhanced training
def run_enhanced_training():
    # Load tokenizer
    tokenizer = AutoTokenizer.from_pretrained('bert-base-multilingual-cased')

    # Initialize enhanced model
    model = EnhancedContrastiveModel().to(device)

    # Load base model weights (where compatible)
    model = load_base_model(model)

    # Load data from Excel files
    tatian_latin, tatian_ohg = load_excel_aligned_texts(
        os.path.join(DATA_DIR, 'Tatian_aligned.xlsx'), 'Tatian')

    isidor_latin, isidor_ohg = load_excel_aligned_texts(
        os.path.join(DATA_DIR, 'Isidor_aligned.xlsx'), 'Isidor')

    physio_latin, physio_ohg = load_excel_aligned_texts(
        os.path.join(DATA_DIR, 'Physiologus_aligned.xlsx'), 'Physiologus')

    benedict_latin, benedict_ohg = load_excel_aligned_texts(
        os.path.join(DATA_DIR, 'Benedictine_Rule_aligned.xlsx'), 'Benedictine Rule')

    # Combine all texts
    all_latin = tatian_latin + isidor_latin + physio_latin + benedict_latin
    all_ohg = tatian_ohg + isidor_ohg + physio_ohg + benedict_ohg

    print(f"Total loaded: {len(all_latin)} parallel texts")

    # Create positive pairs (true translations)
    positive_pairs = list(zip(all_latin, all_ohg))

    # Create stronger negative pairs
    negative_pairs = create_stronger_negative_pairs(all_latin, all_ohg)

    # Split data into training and validation
    # Split positive pairs
    pos_split_idx = int(0.9 * len(positive_pairs))
    train_pos_pairs = positive_pairs[:pos_split_idx]
    val_pos_pairs = positive_pairs[pos_split_idx:]

    # Split negative pairs
    neg_split_idx = int(0.9 * len(negative_pairs))
    train_neg_pairs = negative_pairs[:neg_split_idx]
    val_neg_pairs = negative_pairs[neg_split_idx:]

    print(f"Training set: {len(train_pos_pairs)} positive pairs, {len(train_neg_pairs)} negative pairs")
    print(f"Validation set: {len(val_pos_pairs)} positive pairs, {len(val_neg_pairs)} negative pairs")

    # Create datasets with balanced sampling
    train_dataset = EnhancedContrastiveDataset(train_pos_pairs, train_neg_pairs, tokenizer)
    val_dataset = EnhancedContrastiveDataset(val_pos_pairs, val_neg_pairs, tokenizer)

    # Create data loaders
    batch_size = 16 if device.type == 'cpu' else 32
    train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
    val_loader = DataLoader(val_dataset, batch_size=batch_size)

    # Train the model
    model, train_losses, val_losses = train_with_strong_negatives(
        model,
        train_loader,
        val_loader,
        epochs=5
    )

    # Evaluate the model
    evaluation_results = evaluate_enhanced_model(model, tokenizer)

    return model, evaluation_results

# Interactive testing function
def test_interactive(model, tokenizer):
    """Interactive testing of the model with user input"""
    model.eval()

    print("\n=== Interactive Testing Mode ===")
    print("Enter Latin and OHG text pairs to compute similarity")
    print("Enter 'q' to quit")

    # Higher thresholds based on our enhanced model
    thresholds = {
        'translation': 0.85,
        'paraphrase': 0.70,
        'adaptation': 0.55
    }

    while True:
        print("\n")
        latin = input("Enter Latin text: ")
        if latin.lower() == 'q':
            break

        ohg = input("Enter OHG text: ")
        if ohg.lower() == 'q':
            break

        # Tokenize
        latin_encoding = tokenizer(
            latin,
            return_tensors='pt',
            max_length=128,
            padding='max_length',
            truncation=True
        ).to(device)

        ohg_encoding = tokenizer(
            ohg,
            return_tensors='pt',
            max_length=128,
            padding='max_length',
            truncation=True
        ).to(device)

        # Compute embeddings
        with torch.no_grad():
            latin_embedding = model(latin_encoding['input_ids'], latin_encoding['attention_mask'])
            ohg_embedding = model(ohg_encoding['input_ids'], ohg_encoding['attention_mask'])

        # Compute similarity
        similarity = F.cosine_similarity(latin_embedding, ohg_embedding).item()

        # Determine relationship based on thresholds
        relationship = "unrelated"
        if similarity >= thresholds['translation']:
            relationship = "translation"
        elif similarity >= thresholds['paraphrase']:
            relationship = "paraphrase"
        elif similarity >= thresholds['adaptation']:
            relationship = "adaptation"

        print(f"Similarity: {similarity:.4f}")
        print(f"Relationship: {relationship}")
        print(f"Thresholds: translation ≥ {thresholds['translation']}, paraphrase ≥ {thresholds['paraphrase']}, adaptation ≥ {thresholds['adaptation']}")

# Run enhanced training
if __name__ == "__main__":
    model, evaluation_results = run_enhanced_training()

    # Ask if user wants to test interactively
    test_interactively = input("\nWould you like to test the model interactively? (y/n): ")
    if test_interactively.lower() == 'y':
        tokenizer = AutoTokenizer.from_pretrained('bert-base-multilingual-cased')
        test_interactive(model, tokenizer)

    print("Enhanced training completed!")
```
